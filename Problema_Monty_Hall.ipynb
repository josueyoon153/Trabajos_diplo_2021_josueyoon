```python
import numpy as np
max_train=3000
premio=0
cambio=1
eleccion=2
res=3
#

game = np.array([np.random.randint(0, 3, max_train),np.random.randint(0, 2, max_train),np.random.randint(0, 3, max_train)])
game = np.transpose(game)
```

```python
np.random.randint(0, 3, max_train)
```

```python
for i in game:
  if i[cambio] == 1 : 
    if i[premio] == i[eleccion]: #Si la eleccion inicial es igual al premio
      if i[eleccion] == 0 :
        i[eleccion]= 1 #Elijo cualesquier de las otras opciones
      elif i[eleccion] == 1 :
        i[eleccion] = 0 
      else:
        i[eleccion] = 1
    else:
      #Abro la puerta que no tiene premio, o sea, la elección posible va a ser el premio
      i[eleccion] = i[premio]
```

```python
cambio_acum = 0
cambio_acum_plot = []
no_cambio_acum = 0
no_cambio_acum_plot = []

for i in game:
  if i[eleccion]==i[premio]:
    if i[cambio] == 1:
       cambio_acum=cambio_acum+1

    else:
       no_cambio_acum=no_cambio_acum+1
 
  no_cambio_acum_plot.append(no_cambio_acum)
  cambio_acum_plot.append(cambio_acum)
```

```python
print(cambio_acum, no_cambio_acum)
```

```python
from matplotlib import pyplot as plt
```
